```python
import random
```

```python
def fitness(x,y,z):
    return 3 * x ** 5 + 6 * y ** 7 + 2 * z ** 9 - 46
```

```python
population = []
for _ in range(1000):
    population.append((random.uniform(-1e5,1e5), random.uniform(-1e5,1e5), random.uniform(-1e5,1e5)))
```

```python
answer = ()
for _ in range(5000):
    ranked_solution = []
    for p in population:
        ranked_solution.append((abs(fitness(p[0], p[1], p[2])), p))
    ranked_solution.sort()
    # ranked_solution.reverse()
    best_solution = ranked_solution[:100]
    print(best_solution[0][0])
    answer = best_solution[0][1]
    if abs(best_solution[0][0]) < 0.00001:
        print(best_solution[0][1])
        break
    population = []
    for _ in range(1000):
        a = random.choice(best_solution)[1]
        b = random.choice(best_solution)[1]
        l = 0.9
        r = 1.1
        population.append((a[0] * random.uniform(l, r), 
            b[1] * random.uniform(l, r), 
            a[2] * random.uniform(l, r)))
```
